```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data and drop the frist 4 rows 
df = pd.read_csv("C:/Users/Asus/Desktop/Info teck intern/API_SP.POP.TOTL_DS2_en_csv_v2_3401680/API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv",skiprows=4)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.drop('Unnamed: 68',axis=1)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.fillna(0,inplace=True)
df.isna().sum()
```

```python
df['Total'] = df.loc[:, '1960':'2023'].sum(axis=1)
```

```python
df['Total']
```

```python
plt.figure(figsize=(50, 10))
sns.barplot(x = df['Country Name'],y = df['Total'] , data = df,palette='flare')
plt.xticks(rotation = 90)
plt.title('Country by Total',size=40)
plt.show()
```

```python
df_county_Total = df[['Country Name','Total']]
df_county_Total
```

```python
# Get the top 10 countries by Total
top_10_countries = df_county_Total.nlargest(10, 'Total')
```

```python
top_10_countries
```

```python
sns.barplot(x = top_10_countries['Country Name'],y = top_10_countries['Total'] , data = top_10_countries ,palette='flare')
plt.xticks(rotation = 90)
plt.title('Top 10 Country by Total')
plt.show()
```

```python
years = df.columns[4:-1]

total_values = df[years].sum()

plt.figure(figsize=(30, 30))
plt.barh(years, total_values,color='#191970')
plt.xlabel('Total Values')
plt.ylabel('Year', size=20)
plt.title('Total Values per Year', size=40)
plt.show()
```
